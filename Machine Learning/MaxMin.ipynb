* Exemplo:

Tentaremos localizar os minimos locais de $f(x)  = (x+5)^2$ partindo de x = 3.

```python
# Plotando inicialmente
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,100)
plt.figure(figsize=(4,4))
plt.plot(x, (x+5)**2)
plt.show()
```

```python
# Aparentemente o valor minimo é algo perto de -5. 
# Tentaremos verificar isso aplicando o Gradiente Descendente.

initial = 3
learning_rate = 0.1
iterations = 0
max_interations = 100
epsilon = 1e-6
def function(x):
    return (x + 5)**2

def dev_function(x):
    return 2*(x+5)
```

```python
# Se verificarmos o valor inicial na função definida:
function(initial)
```

```python
# Aparentemente temos um valor elevado... aplicaremos um método iterativo
x = initial
previous = 1
while previous > epsilon and iterations < max_interations:
    prev_x = x
    x = x - learning_rate*dev_function(prev_x)
    iterations = iterations + 1

    previous = abs(x-prev_x)

print(f'Numero de iterações: {iterations} - Valor de x: {x:.4f}')
```
